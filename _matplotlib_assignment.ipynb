```python
# 1. Create a simple line plot for the following data:
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

# a. Plot the data.
# plt.plot() creates a line plot
plt.plot(x, y)

# b. Customize the plot by adding a title, axis labels, and a grid.
# Add a title to the plot
plt.title("Simple Line Plot")

# Add a label to the x-axis
plt.xlabel("X-axis Label")

# Add a label to the y-axis
plt.ylabel("Y-axis Label")

# Add a grid to the plot for better readability
plt.grid(True)

# Display the plot
# plt.show() is essential to display the plot when running as a script,
# though in Jupyter notebooks, plots are often displayed automatically.
plt.show()
```

```python
# 2. Create a bar graph to represent the marks scored by students in a subject:

students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

# a. Plot the data as a bar graph.
# plt.bar() creates a bar graph.
# We can specify colors directly in the bar function.
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold']
plt.bar(students, marks, color=colors)

# b. Customize the colors and add a title.
# Add a title to the bar graph
plt.title("Student Marks in a Subject")

# Add labels for clarity
plt.xlabel("Students")
plt.ylabel("Marks Scored")

# Display the bar graph
plt.show()
```

```python
# 3. Create a pie chart to represent the percentage distribution of a company’s revenue from different regions:

regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# a. Create a pie chart with the region names as labels.
# b. Highlight the region with the highest revenue.

# Determine which slice to "explode" (pull out)
# Find the index of the region with the highest revenue
max_revenue_index = revenue.index(max(revenue))

# Create an 'explode' tuple with a small offset for the highest revenue slice
# The length of explode must be the same as the number of wedges.
explode = [0] * len(regions) # Initialize with zeros
explode[max_revenue_index] = 0.1 # Explode the slice with the highest revenue

# Create the pie chart
# autopct='%1.1f%%' displays the percentage on each slice
# startangle=90 rotates the start of the first slice
# explode highlights a slice
plt.pie(revenue, labels=regions, autopct='%1.1f%%', startangle=90, explode=explode)

# Add a title to the pie chart
plt.title("Company Revenue Distribution by Region")

# Ensure the pie chart is drawn as a circle.
plt.axis('equal')

# Display the pie chart
plt.show()
```

```python
# 4. Generate a histogram to show the frequency distribution of randomly generated integers
# between 1 and 100 (sample size = 1000).

# Generate 1000 random integers between 1 and 100 (inclusive)
# np.random.randint(low, high, size) generates integers from low (inclusive) to high (exclusive)
# So, for 1 to 100 inclusive, we use 1 and 101.
import numpy as np
random_data = np.random.randint(1, 101, 1000)

# Create the histogram
# bins specifies the number of bins or the bin edges
# edgecolor adds a border to the bars for better visibility
plt.hist(random_data, bins=20, edgecolor='black', alpha=0.7) # Using 20 bins for better distribution visualization

# Add a title and axis labels
plt.title("Frequency Distribution of Random Integers (1-100)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Add a grid for better readability
plt.grid(axis='y', alpha=0.75)

# Display the histogram
plt.show()
```
